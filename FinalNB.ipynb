# Working on Dataset by building Naive Bayes Classifier

Import Breast Cancer Dataset from UCI Repository

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import preprocessing
from sklearn.datasets import load_breast_cancer
from datetime import datetime
```

Creating a DataFrame with assigned Features and Target Values

```python
#Naive Bayes
#Start here

X = load_breast_cancer()['data']
y = load_breast_cancer()['target']
names = load_breast_cancer()['feature_names']


full_df = pd.DataFrame(X, columns = names)
full_df['target'] = y
dtTemp = full_df.iloc[:,:-1]
x = full_df.iloc[:,:-1].values
min_max_scaler = preprocessing.MinMaxScaler(feature_range = (0,2))
x_scaled = min_max_scaler.fit_transform(x)
NormalizedDT = pd.DataFrame(x_scaled,columns=dtTemp.columns)
full_df = NormalizedDT
full_df['target'] = y
```

Importing written KNN classifier code :: 
Importing written Cross Validation classifier code

```python
from NaiveBayes import Naive_Bayes_Classifier, Naive_Bayes_Multiple
from CrossValidation import Train_Test_Split, Classifier, KFold_CV 
```

Create a Training and Testing DataFrame 

```python
split_data = Train_Test_Split(full_df)
```

Create List of predicted values from NB Classifier with a timeframe

```python
y_pred_list = []

for fold in split_data:

    now = datetime.now()
    dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
    print("date and time =", dt_string)
    
    y_pred = Naive_Bayes_Multiple(fold[0], np.array(fold[1].iloc[:,:-1]))
    y_pred_list.append(y_pred)
    
    now = datetime.now()
    dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
    print("date and time =", dt_string)
```

Create a list of Test target values 

```python
y_test_list = []
for fold in split_data:
    y_test = list(fold[1].iloc[:,-1].values)
    y_test_list.append(y_test)
```

Check accuracy by comparing test and predicted target values by calling KFOLD Function

```python
KFold_CV(y_test_list,y_pred_list,full_df)
```
